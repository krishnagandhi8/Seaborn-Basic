## Plotting categorical scatter

```python
import matplotlib.pyplot as plt 
import seaborn as sns 
%matplotlib inline  
# x axis values 
x =['sun', 'mon', 'fri', 'sat', 'tue', 'wed', 'thu'] 
  
# y axis values 
y =[5, 6.7, 4, 6, 2, 4.9, 1.8] 
  
# plotting strip plot with seaborn 
ax = sns.stripplot(x, y); 
  
# giving labels to x-axis and y-axis 
ax.set(xlabel ='Days', ylabel ='Amount_spend') 
  
# giving title to the plot 
plt.title('Categorical Scatter'); 
  
# function to show plot 
plt.show() 
```

## Stripplot 

```python
sns.set(style ="whitegrid") 
  
# loading data-set 
iris = sns.load_dataset('iris'); 

ax = sns.stripplot(x = 'species', y = 'sepal_length', data = iris); 
  
# giving title to the plot 
plt.title('Graph') 
  
# function to show plot 
plt.show() 
```

## Swarm Plot 

```python
ax = sns.swarmplot(x = 'species', y = 'sepal_length', data = iris); 
  
# giving title to the plot 
plt.title('Graph') 
  
# function to show plot 
plt.show()
```

## Swarmplot 

```python
ax = sns.swarmplot(x = 'sepal_length', y = 'species', data = iris); 
  
plt.title('Graph') 
  
# function to show plot 
plt.show() 
```

## Relationship graph 

```python
sns.set()
tips = sns.load_dataset("tips")
sns.relplot(x="total_bill", y="tip", col="time",
            hue="smoker", style="smoker", size="size",
            data=tips);
```

```python
dots = sns.load_dataset("dots")
sns.relplot(x="time", y="firing_rate", col="align",
            hue="choice", size="coherence", style="choice",
            facet_kws=dict(sharex=False),
            kind="line", legend="full", data=dots);
```

## Specialized categorical plots 

```python
sns.catplot(x="day", y="total_bill", hue="smoker",
            kind="swarm", data=tips);
```

```python
sns.catplot(x="day", y="total_bill", hue="smoker",
            kind="violin", split=True, data=tips);
```

```python
sns.catplot(x="day", y="total_bill", hue="smoker",
            kind="bar", data=tips);
```

## Boxplot and Scatter 

```python
f, axes = plt.subplots(1, 2, sharey=True, figsize=(6, 4))
sns.boxplot(x="day", y="tip", data=tips, ax=axes[0])
sns.scatterplot(x="total_bill", y="tip", hue="day", data=tips, ax=axes[1]);
```

## Joint 

```python
iris = sns.load_dataset("iris")
sns.jointplot(x="sepal_length", y="petal_length", data=iris);
```
